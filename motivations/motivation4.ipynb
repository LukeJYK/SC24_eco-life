## Test with different co-optimize parameters

```python
fig, axs = plt.subplots(nrows=1, ncols=1, gridspec_kw={'hspace': 0.4, 'wspace': 0.1, 'bottom': 0.2, 
                    'top': 0.8, 'right':0.995, 'left':0.17}, figsize=(6.5,3), sharey=True)
FONTSIZE =13
XLABEL = "CO$_2$ Footprint \n (% increase w.r.t. CO$_2$-Opt)"
YLABEL = "Service Time (%\n increase w.r.t.\nService-Time-Opt)"

carbon_opt = [3.516255909582706, 0.018386601881651327]
oracle = [2.7717964527099626,0.019433615515948306, ]
perf_opt =[2.6050180581418347, 0.026339604021742274]
energy_opt=[3.778154377851147,0.02282129780386102]
x_move = 0
y_move =0
min_st = perf_opt[0]
min_carbon = carbon_opt[1]
carbon_opt_percent = [100*(carbon_opt[0]-min_st)/min_st+y_move,100*(carbon_opt[1]-min_carbon)/min_carbon+x_move+1]
oracle_percent = [100*(oracle[0]-min_st)/min_st+y_move,100*(oracle[1]-min_carbon)/min_carbon+x_move]
energy_opt_percent = [100*(energy_opt[0]-min_st)/min_st+y_move,100*(energy_opt[1]-min_carbon)/min_carbon+x_move]
perf_opt_percent =[100*(perf_opt[0]-min_st)/min_st+y_move+3,100*(perf_opt[1]-min_carbon)/min_carbon+x_move]

x = [carbon_opt_percent[1],oracle_percent[1],perf_opt_percent[1],energy_opt_percent[1]]
y = [carbon_opt_percent[0],oracle_percent[0],perf_opt_percent[0], energy_opt_percent[0]]
axs.set_xlabel(XLABEL,fontsize=FONTSIZE)
axs.set_ylabel(YLABEL,fontsize=FONTSIZE)
axs.tick_params(axis='both', which='major', pad=1, labelsize=FONTSIZE)
axs.grid(which='both', color='lightgrey', ls='dashed', zorder=0)
colors = ['#7fc97f','#DAA520','#beaed4','#FF69B4']

markers = ['v', 'X', 's','D']
LABELS = ['CO$_2$-Opt','Oracle','Service-Time-Opt',"Energy-Opt"]
for i in range(len(x)):
    axs.scatter(x=x[i], y=y[i], color=colors[i], label=LABELS[i], s=200, zorder=3, alpha=1, edgecolors="black", marker=markers[i])

axs.legend(loc=(-0.05,1.03), frameon = False ,ncol=4,labels=LABELS,fontsize=13,columnspacing=0.4,labelspacing = 1)
axs.yaxis.set_major_formatter(mtick.FormatStrFormatter('%i'))
axs.set_ylim(ymin=0,ymax=50)
axs.set_xlim(xmin=0,xmax=50)
plt.savefig("motiv4.pdf",bbox_inches='tight')
```
